### Import Dependencies

```python
from langsmith import Client
```

### Attach Feedback to a specific trace_id in Langsmith

d1ff45a8-debb-4dda-9d4c-751db02e92e3

```python
client = Client()
```

```python
client.create_feedback(
    run_id="d1ff45a8-debb-4dda-9d4c-751db02e92e3",
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id="d1ff45a8-debb-4dda-9d4c-751db02e92e3",
    key="feedback_1",
    score=1,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id="d1ff45a8-debb-4dda-9d4c-751db02e92e3",
    key="comment",
    value="This is a long form feedback you should use to improve the system!",
    feedback_source_type="api"
)
```
